```python
#reviews ---> [eating,eat,eaten],[going,goes,gone]
words=["eating","eat","eaten","going","goes","gone","went" ,"writes","history","writing","written"]
print(words)
```

```python
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
for     word in words:
    print(f" {word} -->  {stemming.stem(word)}")
```

```python
stemming.stem('congratulations')
```

```python
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$',min=4)
reg_stemmer.stem("running")
reg_stemmer.stem("ingswimming")
```

Snowball Stemmer

work better than porter stemmmer

```python
from nltk.stem import SnowballStemmer
snow_stemmer=SnowballStemmer('english')
for word in words:
    print(f" {word} -->  {snow_stemmer.stem(word)}")
```

```python
stemming.stem("fairly"),snow_stemmer.stem("fairly")
```
