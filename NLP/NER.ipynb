```python
sentence="Amazon is expanding rapidl . The Amazon is the largest rainforest "
```

```python
from nltk.tokenize import word_tokenize
tokens=word_tokenize(sentence)
tokens
```

```python
import nltk
```

```python
tag_elements=nltk.pos_tag(tokens)
tag_elements
```

```python
nltk.download('maxent_ne_chunker_tab')
```

```python
nltk.download('words')
```

```python
nltk.ne_chunk(tag_elements).draw()
```
