WordNet Lemmatizer-lemmas to root word/dictionary word

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
words=["eating","eat","eaten","going","goes","gone","went" ,"writes","history","writing","written"]
for word in words:
    print(f" {word} -->  {lemmatizer.lemmatize(word,pos='v')}")
```
