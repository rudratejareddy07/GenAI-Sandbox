```python
corpus="I am Rudra , a love to learn new technologies ! I love to code and create new projects."
print(corpus)
```

```python

import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
documents=sent_tokenize(corpus)
for sentence in documents:
    print(sentence)
```

```python
#paragraph to sentences/words
from nltk.tokenize import sent_tokenize, word_tokenize
documents=word_tokenize(corpus)
for word in documents:
    print(word)
```
